```python
folder = "results" # original survey results
models = "_".join(systems)
final_path = os.path.join(folder,models)
df2 = pd.read_csv(final_path)
df2
```

```python

# process the co-occurance of both emotions within the same situation
fear_happiness = (df2['fear of troubling someone else'] > 0) & (df2['happiness'] > 0)
positive_negative = (df2['positive'] > 0) & (df2['negative'] > 0)

df2.loc[:,("fear_happiness")] = fear_happiness
df2.loc[:,("positive_negative")] = positive_negative
```

```python
# process the co-occurance of both emotions within the same situation
fear_happiness = (df['fear of troubling someone else'] > 0) & (df['happiness'] > 0)
positive_negative = (df['positive'] > 0) & (df['negative'] > 0)

df.loc[:,("fear_happiness")] = fear_happiness
df.loc[:,("positive_negative")] = positive_negative
```

```python
# t-test all conditions to find similarities
from statsmodels.stats.weightstats import ttest_ind
from itertools import combinations

emotion_dict = {'self-success':['motivation', 'responsible for others','fear_happiness', 'positive_negative'],\
                'self-failure': ['motivation', 'responsible for myself', 'responsible for others']}
systems = ["mistral-7b-instruct", "gemma-7b-it:free", "llama-2-70b-chat", "gpt-3.5-turbo", "gpt-4-turbo-preview"]
conditions = ["jp_jp", "jp_en", "en_jp", "en_en", "en", "jp"]
conditions = ["en", "en_en", "en_jp", "jp_en", "jp_jp", "jp"]

culture = ["us", "jp"]
relation = "two-sided"
for system in systems:
    new = pd.DataFrame(0, columns=conditions, index=conditions)
    # extract system's samples
    one = df[df['system']==system]
    two = df2[df2['system']==system]
    # split df to two different poulations us and jp
    k=0
    for stype in emotion_dict.keys():
        stype_df = one[one['status']==stype]
        stype_df2 = two[two['status']==stype]
        
        for emotion in emotion_dict[stype]:
            for condition1, condition2 in combinations(conditions, 2):

                if condition1 == "us":
                    c1 = stype_df2[stype_df2['culture']=='us']
                elif condition1 == "jp":
                    c1 = stype_df2[stype_df2['culture']=='japanese']
                else:
                    c1 = stype_df[stype_df['condition']==condition1]

                if condition2 == "us":
                    c2 = stype_df2[stype_df2['culture']=='us']
                elif condition2 == "jp":
                    c2 = stype_df2[stype_df2['culture']=='jp']
                else:
                    c2 = stype_df[stype_df['condition']==condition2]

                c1_emotion = c1[emotion]
                c2_emotion = c2[emotion]
                # generate a dataframe
                _, pvalue, _ = ttest_ind(c1_emotion, c2_emotion, alternative=relation)
                if pvalue > 0.05:
                    new.loc[condition1, condition2]+=1
                    new.loc[condition2, condition1]+=1
                
    print(system)
    print(new)

```
