```python
import pandas as pd
import os

folder = "results" # original survey results
models = "_".join(systems)
final_path = os.path.join(folder,models)
df2 = pd.read_csv(final_path)
df2
```

```python
# Sample n=80
N=800 # each participant responds to 10 situations (which are 10 rows x 80 participants)
us_df = df2[df2['culture']=='us']
jp_df = df2[df2['culture']=='japanese']
df_us = us_df.groupby('system').head(N).reset_index(drop=True)
df_jp = jp_df.groupby('system').head(N).reset_index(drop=True)
df = pd.concat([df_us,df_jp])
df = df.drop(['Unnamed: 0'], axis=1)
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
# self success Appraisals and Coping strategies (test 1, and 2)
from statsmodels.stats.weightstats import ttest_ind
emotions = ["motivation", "responsible for others"]
self_success_df = df[df['status']=="self-success"]
systems = ["mistral-7b-instruct", "gemma-7b-it:free", "llama-2-70b-chat", "gpt-3.5-turbo", "gpt-4-turbo-preview"]



relations = ['larger', 'larger'] # reported relationships
relations = ['smaller', 'smaller'] # opposite from reported relationships

for system in systems:
    one = self_success_df[self_success_df['system']==system]
    us = one[one['culture']=='us']
    jp = one[one['culture']=='japanese']
    for (relation,emotion) in zip(relations, emotions):
        jp_m = jp[emotion].mean()
        us_m = us[emotion].mean()
        jp_std = jp[emotion].std()
        us_std = us[emotion].std()
        jp_emotion = jp[emotion]
        us_emotion = us[emotion]
        _, pvalue, _ = ttest_ind(jp_emotion, us_emotion, alternative=relation)
        if pvalue < 0.05:
            print(f'H0 is rejected for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
        else:
            print(f'H0 is accepted for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
```

```python
# self failure Appraisals and Coping strategies (test 5, 6, and 7)
from statsmodels.stats.weightstats import ttest_ind
emotions = ["responsible for others", "responsible for myself", "motivation"]
self_failure_df = df[df['status']=="self-failure"]
systems = ["mistral-7b-instruct", "gemma-7b-it:free", "llama-2-70b-chat", "gpt-3.5-turbo", "gpt-4-turbo-preview"]


relations = ['larger','larger', 'smaller'] # reported relationships
relations = ['smaller','smaller', 'larger'] # opposite from reported relationships

for system in systems:
    one = self_failure_df[self_failure_df['system']==system]
    us = one[one['culture']=='us']
    jp = one[one['culture']=='japanese']
    for (relation,emotion) in zip(relations, emotions):
        jp_m = jp[emotion].mean()
        us_m = us[emotion].mean()
        jp_std = jp[emotion].std()
        us_std = us[emotion].std()
        jp_emotion = jp[emotion]
        us_emotion = us[emotion]
        _, pvalue, _ = ttest_ind(jp_emotion, us_emotion, alternative=relation)
        if pvalue < 0.05:
            print(f'H0 is rejected for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
        else:
            print(f'H0 is accepted for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
```

```python
# Co-occurance of Positive and Negative emotions (test 3, and 4)
self_success_df = df[df['status']=="self-success"]
# Create a new colum to indicate whether both emotions were triggered at the same time
# pair 1 positive and negative emotions
# pair 2 happy and fear of troubling others
fear_happiness = (self_success_df['fear of troubling someone else'] > 0) & (self_success_df['happiness'] > 0)
positive_negative = (self_success_df['positive'] > 0) & (self_success_df['negative'] > 0)

self_success_df.loc[:,("fear_happiness")] = fear_happiness
self_success_df.loc[:,("positive_negative")] = positive_negative
systems = ["mistral-7b-instruct", "gemma-7b-it:free", "llama-2-70b-chat", "gpt-3.5-turbo", "gpt-4-turbo-preview"]


emotions = ["fear_happiness", "positive_negative"]
# per system compare us jp
relation = "larger" # reported relationships
relation = "smaller" # opposite from reported relationships
for emotion in emotions:
    for system in systems:
        one = self_success_df[self_success_df['system']==system]
        us = one[one['culture']=='us']
        jp = one[one['culture']=='japanese']
        emotion_us = us[emotion]
        emotion_jp = jp[emotion]
        _, pvalue, _ = ttest_ind(emotion_jp, emotion_us, alternative=relation)
        if pvalue < 0.05:
            print(f'H0 is rejected for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
        else:
            print(f'H0 is accepted for {system} with p-value of {pvalue} on Japanese exhibiting {relation} {emotion} than Americans')
```
