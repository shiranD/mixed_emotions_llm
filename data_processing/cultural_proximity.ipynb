```python
# construct a df across all 8 languages and compare results across tests
import pandas as pd
import os

folder = "results/languages"
systems = ['mistral', 'gemma', 'llama', 'gpt3.5', 'gpt4']
languages = ["vt", "kr", "ch", "jp", "gr", "fr", "sp", "en"]


l =  # n number of samples
frames = pd.DataFrame()
path = []
for language in languages:
    for system in systems:
        for i in range(l):
            path = os.path.join(folder, system, language+"_"+system+str(i))
            #print(path)
            df = pd.read_csv(path)
            frames = pd.concat([frames, df])
models = "_".join(systems)
print(os.path.join(folder,models))
final_path = os.path.join(folder,models)
frames = frames.drop(['Unnamed: 0'], axis=1)
frames = frames.reset_index()
frames = frames.drop(['index'], axis=1)
frames.to_csv(final_path)
frames
```

```python
folder = "results" # original survey results
models = "_".join(systems)
final_path = os.path.join(folder,models)
df2 = pd.read_csv(final_path)
df2
```

```python
df = pd.concat([frames, df2])
df = df.reset_index()
df = df.drop(['index'], axis=1)
df
```

```python
# preparing the df with for all tests
fear_happiness = (df['fear of troubling someone else'] > 0) & (df['happiness'] > 0)
positive_negative = (df['positive'] > 0) & (df['negative'] > 0)

df.loc[:,("fear_happiness")] = fear_happiness
df.loc[:,("positive_negative")] = positive_negative
```

```python
# crisscross all conditions to find similarities
from statsmodels.stats.weightstats import ttest_ind
from itertools import combinations

emotion_dict = {'self-success':['motivation', 'responsible for others','fear_happiness', 'positive_negative'],\
                'self-failure': ['motivation', 'responsible for myself', 'responsible for others']}
systems = ["mistral-7b-instruct", "gemma-7b-it:free", "llama-2-70b-chat", "gpt-3.5-turbo", "gpt-4-turbo-preview"]
languages = ["vt", "kr", "ch", "jp", "gr", "fr", "sp", "en"]
languages = ["vietnamese", "korean", "chinese", "japanese", "french", "spanish", "english"]


relation = "two-sided"
for system in systems:
    # extract system's samples
    one = df[df['system']==system]
    # split df to two different poulations us and jp
    for stype in emotion_dict.keys():
        stype_df = one[one['status']==stype]
        new = pd.DataFrame(0, columns=languages, index=languages)
        for emotion in emotion_dict[stype]:
            for condition1, condition2 in combinations(languages, 2):
                c1 = stype_df[stype_df['culture']==condition1]
                c2 = stype_df[stype_df['culture']==condition2]
                c1_emotion = c1[emotion]
                c2_emotion = c2[emotion]
                # generate a dataframe
                _, pvalue, _ = ttest_ind(c1_emotion, c2_emotion, alternative=relation)
                if pvalue > 0.05:
                    new.loc[condition1, condition2]+=1
                    new.loc[condition2, condition1]+=1


                
    print(system)
    print(new)
```
